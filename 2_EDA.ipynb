```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
import matplotlib.patches as patches
```

```python
#load the data
file_path = "C:\data\lap_time.csv"
lap_data = pd.read_csv(file_path)
```

```python
#basic info about dataset
lap_data_info = {
    "Shape": lap_data.shape,
    "Columns": lap_data.columns.tolist(),
    "Missing Values": lap_data.isnull().sum(),
    "Data Types": lap_data.dtypes
}

lap_data_info
```

```python
import matplotlib.pyplot as plt

# lap per driver in season
laps_per_driver = lap_data['Driver'].value_counts()

#lap times
lap_time_distribution = lap_data['LapTimeSeconds']

#tyres used
compound_usage = lap_data['Compound'].value_counts()

#pitstops per driver
stints_per_driver = lap_data.groupby('Driver')['Stint'].nunique()

#lap time per year on avg
yearly_lap_times = lap_data.groupby('Year')['LapTimeSeconds'].mean()

#trackstatus
track_status_counts = lap_data['TrackStatus'].value_counts()
```

```python
#plot total laps per driver
plt.figure(figsize=(14, 6))
laps_per_driver.plot(kind='bar')
plt.title('Total Laps Per Driver')
plt.xlabel('Driver')
plt.ylabel('Number of Laps')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#plot lap time distrubtion
plt.figure(figsize=(10, 5))
lap_time_distribution.hist(bins=50)
plt.title('Lap Time Distribution')
plt.xlabel('Lap Time (secs)')
plt.ylabel('Frequency')
plt.show()
```

```python
#tyre compound plot
plt.figure(figsize=(8, 5))
compound_usage.plot(kind='bar')
plt.title('Tyre Compound Usage')
plt.xlabel('Compound')
plt.ylabel('Number of Laps')
plt.tight_layout()
plt.show()
```

```python
#stint plot
plt.figure(figsize=(14, 6))
stints_per_driver.plot(kind='bar', color='orange')
plt.title('Number of Stints per Driver')
plt.xlabel('Driver')
plt.ylabel('Number of Unique Stints')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#avg lap time plot
plt.figure(figsize=(8, 5))
yearly_lap_times.plot(marker='o')
plt.title('Average Lap Time per Year')
plt.xlabel('Year')
plt.ylabel('Average Lap Time (seconds)')
plt.grid(False)
plt.tight_layout()
plt.show()
```
